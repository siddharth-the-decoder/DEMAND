```python
prep.dataset['key'] = prep.df['week'].astype(str) + '_' + prep.df['store_id'].astype(str)
```

```python
prep.dataset = prep.df.drop(['record_ID', 'week', 'store_id', 'sku_id', 'total_price', 'base_price', 'is_featured_sku', 'is_display_sku'], axis=1)
```

```python
prep.dataset = prep.df.groupby('key').sum()
```

```python
prep.df
```

```python
prep.df[:100].plot(figsize=(12,8))
```

```python
prep.df['day_1'] = prep.df['units_sold'].shift(-1)
prep.df['day_2'] = prep.df['units_sold'].shift(-2)
prep.df['day_3'] = prep.df['units_sold'].shift(-3)
prep.df['day_4'] = prep.df['units_sold'].shift(-4)
```

```python
prep.df
```

```python
df = prep.df.dropna()
```

```python
df[:100].plot(figsize=(12,8))
```

```python
x1, x2, x3, x4, y = df['day_1'], df['day_2'], df['day_3'], df['day_4'], df['units_sold']
x1, x2, x3, x4, y = np.array(x1), np.array(x2), np.array(x3), np.array(x4), np.array(y)
x1, x2, x3, x4, y = x1.reshape(-1,1), x2.reshape(-1,1), x3.reshape(-1,1), x4.reshape(-1,1), y.reshape(-1,1)

split_percentage = 15
test_split = int(len(df)*(split_percentage/100))
x = np.concatenate((x1, x2, x3, x4), axis=1)
X_train,X_test,y_train,y_test = x[:-test_split],x[-test_split:],y[:-test_split],y[-test_split:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train, y_train)

y_pred = rf_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Predict the values using the test set
y_pred = rf_regressor.predict(X_test)

# Calculate R score
r2 = r2_score(y_test, y_pred)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Print the results
print(f"R Score: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
import random
```

```python

# Function to take input from the farmer and predict the units sold for the next day
def predict_units_sold():
    try:
        # Taking input from the user (farmer) for the past four days
        day_1 = float(input("Enter sales for day 1: "))
        day_2 = float(input("Enter sales for day 2: "))
        day_3 = float(input("Enter sales for day 3: "))
        day_4 = float(input("Enter sales for day 4: "))

        # Creating input array for the model
        input_data = np.array([[day_1, day_2, day_3, day_4]])

        # Making prediction using the trained model
        prediction = rf_regressor.predict(input_data)[0]

        # Apply random loss of up to 10%
        random_loss_percentage = random.uniform(0, 0.10)  # Random loss between 0 and 10%
        final_prediction = prediction * (1 - random_loss_percentage)

        # Display the prediction
        print(f"Predicted units sold for the next day (after 10% random loss): {final_prediction:.2f}")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

# Call the function to take input from the user and predict units sold for the next day
predict_units_sold()
```
